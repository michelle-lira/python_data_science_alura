## Utilizando o método isin()

__O que faz o método isin()?__

Passando uma lista com os tipos de imóveis, ele vai verificar dentro do DataFrame   
e se essa variável tipo tiver um imóvel que está na lista, ele vai marcar este registro     
como verdadeiro, senão, vai retornar falso.

```python
# Checando mais uma vez esta o DataFrame, pra poder fazer uma comparação depois de usar o isin()
dados.head(10)
```

```python
# Criando uma Series booleana indicando os elementos presentes na seleção de imóveis residenciais
dados['Tipo'].isin(residencial).head(10)
```

```python
# Armazenando a seleção de imóveis residenciais em uma variável
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
# Criando um DataFrame com as informações selecionadas
dados_residenciais = dados[selecao]
```

```python
dados_residenciais
```

```python
# Verificando se foram mantidos apenas os dados selecionados
list(dados_residenciais['Tipo'].drop_duplicates())
```

```python
# Checando o tamanho do DataFrame, quantos dados tem nele
dados_residenciais.shape[0]
```

```python
# Checando o DataFrame criado anteriormente, para comparar
dados.shape[0]
```

```python
# Reconstruindo o index
dados_residenciais.index = range(dados_residenciais.shape[0])
```

```python
dados_residenciais
```

## 01 - Determinando seleções

Veja o código que cria um DataFrame de letras e números.

```python
import pandas as pd

numeros = [i for i in range(11)]
letras = [chr(i + 65) for i in range (11)]
nome_coluna = ['N']

df = pd.DataFrame(data = numeros, index = letras, columns = nome_coluna)
df
```

E baseado nesses DataFrame geramos uma seleção para filtrar os dados:

```python
selecao = df['N'].isin([i for i in range(11) if i % 2 == 0])
df = df[selecao]
df
```
