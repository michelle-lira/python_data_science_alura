## <font color='purple'>Criando Novas Variáveis</font>

#### Recebemos uma nova demanda do nosso chefe na seguradora:
__Precisamos criar novas variáveis para o conjunto de dados.__    
1. Um das variáveis abrigará o valor bruto do aluguel,    
2. outra que armazena o valor com base no metrô quadrado de um imóvel e, por fim,   
3. uma nova variável de tipo que agregue casas e apartamentos.

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
# Criando a nova variável 'Valor Bruto':
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
# Criando a nova variável 'Valor por m²':
dados['Valor por m²'] = (dados['Valor'] / dados['Area'])
```

```python
dados.head(10)
```

```python
# Reduzindo a quantidade de casas decimais da variável 'Valor por m²'
dados['Valor por m²'] = dados['Valor por m²'].round(2)
```

```python
dados.head(10)
```

```python
# Criando a nova variável 'Valor Bruto m²':
dados['Valor Bruto m²'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

```python
# Criando a nova variável de Tipo que agregue Casas e Apartamentos:

# Primeiro vamos criar uma variável que contenha todos os tipos de casas:
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

## Utilizando o método apply()

__Como o apply() funciona?__    
Permite que nós apliquemos uma função a cada registro ou linha do DataFrame,
no nosso caso, a cada registro.

__Funções lambda:__   
São funções que o usuário não precisa definir, ou seja, não vai precisar escrever
a função para só depois utilizá-la dentro do código.

```python
# Em seguida vamos usar o método apply e lambda
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')

"""
Lê-se a função lambda da seguindo forma:
Cada registro dentro da variável 'Tipo', se estiver especificado na variável casa, criada anteriormente,
será nomeado como Casa, se não, será nomeado como Apartamento.
"""
```

## Excluindo Variáveis

```python
dados_aux.head(10)
```

```python
# Forma de exclusão 1: usando o del
del dados_aux['Valor Bruto']
```

```python
dados_aux.head(10)
```

```python
# Forma de exclusão 2: usando o pop()
dados_aux.pop('Valor Bruto m²')
```

```python
dados_aux.head(10)
```

```python
# Forma de exclusão 3: usando o drop()
dados.drop(['Valor Bruto', 'Valor Bruto m²'], axis = 1, inplace = True)

# Lembrar de colocar inplace = True, senão não vai funcionar
```

```python
dados.head(10)
```
